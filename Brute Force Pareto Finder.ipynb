```python
import numpy as np
import pandas as pd
```

```python
eat_lots_numbers = np.arange(1.000, 1.360, .005)
eat_little_numbers = np.arange(.780, 1.000, .005)
eat_little_little_numbers = np.arange(.700, .780, .005)

lots_price1 = 23.2845
lots_price2 = 33.34283333
lots_price3 = 20.0445
little_price1 = 4.352
little_price2 = 0
little_little_price = 19.24616667

total = 100.27
```

```python
lots_numbers = []
little_numbers = []
little_little_numbers = []
modified_totals = []
for i in eat_lots_numbers:
    lots_modified = (lots_price1 + lots_price2 + lots_price3) * i
    for j in eat_little_numbers:
        little_modified = (little_price1 + little_price2) * j
        for l in eat_little_little_numbers:
            little_little_modified = (little_little_price) * l
            modified_total = lots_modified + little_modified + little_little_modified
            modified_totals.append(modified_total)
            lots_numbers.append(i)
            little_numbers.append(j)
            little_little_numbers.append(l)
```

```python
distances = []
for i in np.arange(len(modified_totals)):
    distances.append(abs(total-modified_totals[i]))
    
multipliers = []
for i in np.arange(len(little_numbers)):
    multipliers.append(lots_numbers[i]/little_numbers[i])
    
multipliers2 = []
for i in np.arange(len(little_little_numbers)):
    multipliers2.append(lots_numbers[i]/little_little_numbers[i])
    
multipliers3 = []
for i in np.arange(len(little_little_numbers)):
    multipliers3.append(little_numbers[i]/little_little_numbers[i])
```

```python
d = {'Lots_Modifier':lots_numbers, 'Little_Modifier':little_numbers, 'Lil_Modifier':little_little_numbers, 'Modified_Totals':modified_totals, 'Absolute_Distance':distances, 'Big-Little':multipliers, 'Big-Lil':multipliers2, 'Little-Lil':multipliers3}
Pareto_Table = pd.DataFrame(data=d)
Pareto_Table.sort_values('Absolute_Distance').head()
```
